```python
import time
import numpy as np
from matplotlib import pyplot
from matplotlib import pyplot as plt
import pickle
import pandas as pd
```

```python
data=pickle.load(open('processed_data.p', "rb" ))
```

```python
seat_data=data.loc[data['seatName']=='B5_9_1_05']
seat_data=seat_data.sort_values(['Datetime'])
seat_data=seat_data.drop(['Country','Location','Resource Type','floor','Department','seatName'],axis=1)
seat_data.set_index('Datetime',inplace=True)
seat_data.index.freq='10T'
```

```python
train,test=seat_data.loc[seat_data.index<'2018-03-25'],seat_data.loc[seat_data.index>='2018-03-25']
```

## SEQUENTIAL SECTION

### Mean 

```python
y_hat_avg = test.copy()
y_hat_avg['avg_forecast'] = train['Space'].mean()
#plt.figure(figsize=(12,8))
plt.plot(train['Space'], label='Train')
plt.plot(test['Space'], label='Test')
plt.plot(y_hat_avg['avg_forecast'], label='Average Forecast')
plt.legend(loc='best')
plt.show()
```

### Moving Average

```python
y_hat_avg=test.copy()
y_hat_avg['moving_avg_forecast'] = train['Space'].rolling(500).mean().iloc[-1]
#plt.figure(figsize=(16,8))
plt.plot(train['Space'], label='Train')
plt.plot(test['Space'], label='Test')
plt.plot(y_hat_avg['moving_avg_forecast'], label='Moving Average Forecast')
plt.legend(loc='best')
plt.show()
```

### Exponential Smoothing

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt

fit2 = ExponentialSmoothing(np.asarray(train['Space'])).fit(smoothing_level=0.6,optimized=False)
y_hat_avg['SES'] = fit2.forecast(len(test))
#plt.figure(figsize=(16,8))
plt.plot(train['Space'], label='Train')
plt.plot(test['Space'], label='Test')
plt.plot(y_hat_avg['SES'], label='SES')
plt.legend(loc='best')
plt.show()
```

### Holt's Linear Trend Method

```python
y_hat_avg = test.copy()
fit1 = Holt(np.asarray(train['Space'])).fit(smoothing_level = 0.3,smoothing_slope = 0.1)
y_hat_avg['Holt_linear'] = fit1.forecast(len(y_hat_avg))
plt.plot(train['Space'], label='Train')
plt.plot(test['Space'], label='Test')
plt.plot(y_hat_avg['Holt_linear'], label='Holt_linear')
plt.legend(loc='best')
plt.show()
```
