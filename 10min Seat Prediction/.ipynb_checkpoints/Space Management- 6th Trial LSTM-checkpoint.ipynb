```python
import numpy as np
from numpy import newaxis
from keras.layers.core import Dense, Activation, Dropout
from keras.layers import CuDNNLSTM
from keras.layers.recurrent import LSTM
from keras.models import Sequential
import pickle
```

```python
data=pickle.load(open('processed_data.p', "rb" ))
trial_data=data.loc[data['seatName']=='B5_9_1_05']
trial_data=trial_data.sort_values(['Datetime'])
#trial_data=trial_data.drop(['Country','Location','Resource Type','floor','Department','seatName','Datetime'],axis=1)
trial_data=trial_data.drop(['Country','Location','Resource Type','floor','Department','seatName'],axis=1)
trial_data.set_index('Datetime',inplace=True)
#trial_data=np.array(trial_data['Space'])
```

```python
train_data,test_data=trial_data.loc[trial_data.index<'2018-03-25'], trial_data.loc[trial_data.index>='2018-03-25']
print(len(train_data),len(test_data))
```

```python
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back):
        a = dataset[i:(i + look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
    print(len(dataY))
    return np.array(dataX), np.array(dataY)
```

```python
look_back = 1
trainX, trainY = create_dataset(train_data.values, look_back)
testX, testY = create_dataset(test_data.values, look_back)
```

```python
trainY
```

```python
trainX.shape,testX.shape
```

```python
trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))
```

```python
model = Sequential()
model.add(LSTM(100, input_shape=(trainX.shape[1], trainX.shape[2])))
model.add(Dense(1))
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
history = model.fit(trainX, trainY, epochs=300, batch_size=100, validation_data=(testX, testY), shuffle=False)
```

```python
from matplotlib import pyplot
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show();
```

```python
pyplot.plot(history.history['acc'], label='train_acc')
pyplot.plot(history.history['val_acc'], label='test_acc')
pyplot.legend()
pyplot.show();
```

```python
yhat = model.predict(testX)
pyplot.plot(yhat, label='predict')
pyplot.plot(testY, label='true')
pyplot.legend()
pyplot.show()
```

```python
predictDates = trial_data.tail(len(testX)).index
yhat = model.predict(testX)
pyplot.plot(predictDates[300:400],yhat[300:400], label='predict')
pyplot.plot(predictDates[300:400],testY[300:400], label='true')
pyplot.legend()
pyplot.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse = sqrt(mean_squared_error(testY, yhat))
print('Test RMSE: %.3f' % rmse)
```
